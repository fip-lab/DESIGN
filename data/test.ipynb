```python
import json

def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data
```

```python
train_aspect = load_json('./train/logs_aspect.json')
train_logs = load_json('./train/logs.json')
count = 0
for log, aspect in zip(train_logs, train_aspect):
    if count >= 120:
        break
    else:
        print(log[-1]["text"])
        print(aspect)
        print()
        count += 1


```
