```python
import json
import random
import torch
import os
import faiss
import numpy as np
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sentence_transformers import SentenceTransformer


import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np


device = torch.device("cuda:1")

dataset_split = "val"

# 加载 tokenizer 和模型
tokenizer = AutoTokenizer.from_pretrained('/home/zhuangjt/zhuangjt_disk3/SK-TOD/results/QuaSiKS_reviewer/checkpoint-78700')  # 指向保存 tokenizer 的目录
model = AutoModelForSeq2SeqLM.from_pretrained('/home/zhuangjt/zhuangjt_disk3/SK-TOD/results/QuaSiKS_reviewer/checkpoint-78700')  # 指向保存模型的目录
model.to(device)

def load_data(path):
    with open(path, 'r') as file:
        data = json.load(file)
    return data

em_knowledge = load_data(f"/home/zhuangjt/zhuangjt_disk3/SK-TOD/pred/{dataset_split}/em/em.json")
logs = load_data(f"/home/zhuangjt/zhuangjt_disk3/SK-TOD/data/{dataset_split}/logs.json")
logs_aspect = load_data(f"/home/zhuangjt/zhuangjt_disk3/SK-TOD/data/{dataset_split}/logs_aspect.json")
vocab_dish_or_drink = load_data("/home/zhuangjt/zhuangjt_disk3/SK-TOD/data/vocab_dish_or_drink.json")

similarity_model = SentenceTransformer('/home/zhuangjt/zhuangjt_disk3/SK-TOD/PLM/all-MiniLM-L6-v2')
##加载资源向量库
resourse_vector = np.load('/home/zhuangjt/zhuangjt_disk3/SK-TOD/InstructKS/Dynamic_InstructKS_Dataset/resourse/resource_vector_base.npy')
dimension = resourse_vector.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(resourse_vector)

resourse_labels = load_data('/home/zhuangjt/zhuangjt_disk3/SK-TOD/InstructKS/Dynamic_InstructKS_Dataset/resourse/labels.json')
resourse_logs = load_data('/home/zhuangjt/zhuangjt_disk3/SK-TOD/InstructKS/Dynamic_InstructKS_Dataset/resourse/logs.json')
resourse_logs_aspect = load_data('/home/zhuangjt/zhuangjt_disk3/SK-TOD/InstructKS/Dynamic_InstructKS_Dataset/resourse/logs_aspect.json')
##先对resourse进行处理，构建三个列表，把所有label["target"] == false的排除掉
resourse_labels_new = []
resourse_logs_new = []
resourse_logs_aspect_new = []
for label,log,log_aspect in zip(resourse_labels,resourse_logs,resourse_logs_aspect):
    if label["target"] == True:
        resourse_labels_new.append(label)
        resourse_logs_new.append(log)
        resourse_logs_aspect_new.append(log_aspect)





##few-shot构建函数
def query_similarity(query_vector):
    
    ##计算query的向量
    D, I = index.search(query_vector, k=1)
    #得到最相似的向量
    similar_vector = resourse_vector[I[0][0]]
    return similar_vector






embeddings1 = []
embeddings2 = []
count = 0
##加载emm_knowledge,logs,logs_aspect
for idx,(em, log, log_aspect) in enumerate(tqdm(zip(em_knowledge, logs, logs_aspect))):
    if count == 150:
        break
    if em["target"] == False:  
        continue
    else:
        user_query = log[-1]["text"]
        query_vector = similarity_model.encode([user_query])
        print(query_vector.shape)
        #query_vector.resize(384,)
        print(query_vector.shape)
        similar_vector = query_similarity(query_vector)
        similar_vector.resize(1, 384)
        print(similar_vector.shape)
        embeddings1.append(query_vector)
        embeddings2.append(similar_vector)
        
        
        count += 1

combined_embeddings = np.vstack((embeddings1, embeddings2))
combined_embeddings = np.squeeze(combined_embeddings, axis=1)
print(combined_embeddings.shape)
# 使用 t-SNE 将合并后的嵌入向量降维到二维空间
tsne = TSNE(n_components=2, random_state=0)
reduced_embeddings = tsne.fit_transform(combined_embeddings)


# 将降维后的嵌入向量分回原来的两组
reduced_embeddings1 = reduced_embeddings[:len(embeddings1), :]
reduced_embeddings2 = reduced_embeddings[len(embeddings1):, :]

# 创建一个散点图，并将对应的点用线连接
plt.figure(figsize=(8, 6))
plt.scatter(reduced_embeddings1[:, 0], reduced_embeddings1[:, 1], c='blue', label='user query embedding')
plt.scatter(reduced_embeddings2[:, 0], reduced_embeddings2[:, 1], c='red', label='most similar embedding')

# 为每一对对应的点画线
for i in range(len(reduced_embeddings1)):
    plt.plot([reduced_embeddings1[i, 0], reduced_embeddings2[i, 0]],
             [reduced_embeddings1[i, 1], reduced_embeddings2[i, 1]], 'gray', linestyle='--')

plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.legend()
plt.grid(True)
plt.show()
```
